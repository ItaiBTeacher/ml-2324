## **עיבוד נתונים מקדים**

```python
# Set ColumnTransformer output to a pandas format
lrn.set_config(transform_output = 'pandas')
```

```python
# Splitting the DataSet to X and y
X = df.drop('loan_status', axis = 1)
y = df['loan_status'].copy()
```

```python
# Split both X and y into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=73)
```

### <span dir=rtl>Column Transformer (מכיל: טיפול בחריגים, טיפול בחסרים, המרת נתונים ונרמול)</span>

```python
# Orginize lists of specific types of data for future ColumnTransformer
numerical_cols = ['person_age',
                  'person_income',
                  'person_emp_length',
                  'loan_amnt',
                  'loan_int_rate',
                  'loan_percent_income',
                  'cb_person_cred_hist_length']

ordinal_cols = ['loan_grade']

categorical_cols = ['person_home_ownership',
                    'loan_intent',
                    'cb_person_default_on_file']
```

```python
# Define the imputation ColumnTransformer parts

imputation_numerical_action = SimpleImputer()
imputation_categorical_action = SimpleImputer(strategy= 'most_frequent')

imputation_parts = [('numerical', imputation_numerical_action, numerical_cols),
                    ('categorial', imputation_categorical_action, categorical_cols+ordinal_cols)]
```

```python
# Define the outliers ColumnTransformer parts

outliers_numerical_action = Winsorizer()

outliers_parts = [('numerical', outliers_numerical_action, numerical_cols)]
```

```python
# Define the categorical to numerial ColumnTransformer parts

cat_to_num_ordinal_action = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
cat_to_num_nominal_action = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

cat_to_num_parts = [('ordinal', cat_to_num_ordinal_action, ordinal_cols),
                    ('nominal', cat_to_num_nominal_action, categorical_cols)]
```

```python
# Define the standardization ColumnTransformer parts

standardization_action = MinMaxScaler()

standardization_parts = [('numericalandordianl', standardization_action, numerical_cols+ordinal_cols)]
```

```python
# Define the each ColumnTransformer

imputation_action = ColumnTransformer(transformers= imputation_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs= -1)
outliers_action = ColumnTransformer(transformers= outliers_parts, remainder= 'passthrough', verbose_feature_names_out=False, n_jobs=-1)
cat_to_num_action = ColumnTransformer(transformers= cat_to_num_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs=-1)
standardization_action = ColumnTransformer(transformers= standardization_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs=-1)
```

```python
# Define the stpes of the preprocessor Pipeline

preprocessing_steps = [('impute', imputation_action),
                       ('outliers', outliers_action),
                       ('cat_to_num', cat_to_num_action),
                       ('standarize', standardization_action)]
```

```python
# Define the preprocessor Pipeline using the preprocessing steps

preprocessor = Pipeline(steps= preprocessing_steps)
```

```python
# Show the preprocessor

preprocessor
```

```python
# Show the data before using ColumnTransformer
X_train
```

```python
# Show the changed the data after using the ColumnTransformer
preprocessor.fit_transform(X_train)
```

## <span dir=rtl>**למידת מכונה**</span>

```python
# Using KNN as a placeholder classifier
placeholder_classifier = KNeighborsClassifier()
```

```python
# Define the classification steps
classification_steps = [('preprocess', preprocessor),
                       ('classifier', placeholder_classifier)]
```

```python
# Define the classifier using the classification steps
clf = Pipeline(steps=classification_steps)
```

```python
# Show the classifier
clf
```

```python
# Create the parameter grid
params = [  {
             'preprocess__standarize__numericalandordianl': [StandardScaler(), MinMaxScaler(), MaxAbsScaler()],
              'classifier': [KNeighborsClassifier()],
              'classifier__n_neighbors': [3, 5, 7, 9]
            },
            {
              'preprocess__standarize__numericalandordianl': [StandardScaler(), MinMaxScaler(), MaxAbsScaler()],
              'classifier': [SVC(random_state=42)],
              'classifier__C': [1, 10, 100, 1000],
              'classifier__kernel': ['linear', 'poly', 'rbf', 'sigmoid']
            }
         ]
```

```python
scorer = make_scorer(accuracy_score)
```

```python
# Define the GridSearchCV
gs = GridSearchCV(estimator = clf,
                  param_grid = params,
                  scoring = scorer,
                  cv=3,
                  n_jobs = -1,
                  verbose= 10,
                  error_score= 'raise')
```

```python
# Show the GridSearchCV
gs
```

```python
gs = helpers.fit(gs, X_train, y_train)
```

```python
# Extract the best parameters from the GridSearchCV
gs.best_params_
```

## **חקר ובחינת ביצועים**

```python
y_pred = helpers.predict(gs, X_test)
```
