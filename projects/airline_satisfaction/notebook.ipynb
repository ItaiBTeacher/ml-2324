<div dir=rtl>

הטבלה אותה בחרתי מגיעה כבר מחולקת לטבלת train ולטבלת test.

**עמודה ראשונה** - נומרית, מספר השורה (עמודת אינדקס)


**Id** - עמודה נומרית, תעודת הזהות של הנוסע

**Gender** - עמודה קטגוריאלית שמית, מגדר הנוסע(Female/Male)

**Customer type** - עמודה קטגוריאלית שמית, סוג הלקוח (Loyal Customer/
Disloyal Customer)

**Age** - עמודה נומרית, גיל הנוסע

**Type of travel** - עמודה קטגוריאלית שמית, סוג הנסיעה (Business travel/personal travel)

**Class** - עמודה קטגוריאלית שמית, מחלקת הנוסע (Bussiness/Eco plus/Eco)

**Flight Distance** - עמודה נומרית, מרחק הטיסה

**Inflight wifi service** - עמודה קטגוריאלית מספרית, דירוג שירות הוויפיי בטיסה מ0 עד 5

**Departure/Arrival time convenient** - עמודה קטגוריאלית מספרית, דירוג נוחות זמן ההמראה והנחיתה מ0 עד 5

**Ease of online booking** - עמודה קטגוריאלית מספרית, דירוג קלות הזמנת הטיסה באינטרנט מ0 עד 5

**Gate location** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של מיקום שער העלייה למטוס

**Food and drink** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של האוכל והשתייה בטיסה

**Online boarding** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של עלייה למטוס

**Seat comfort** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של נוחות המושבים

**Inflight entertainment** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של הבידור בטיסה

**On-board service** - עמודה קטגוריאלית מספרית, דירוג השירות במטוס מ0 עד 5

**Leg room service** - עמודה קטגוריאלית מספרית, דירוג מקום לרגליים במטוס מ0 עד 5

**Baggage handling** - עמודה קטגוריאלית מספרית, דירוג הטיפול במזוודה מ0 עד 5

**Checkin service** - עמודה קטגוריאלית מספרית, דירוג שירות הצ׳ק אין מ0 עד 5

**Inflight service** - עמודה קטגוריאלית מספרית, דירוג השירות בטיסה מ0 עד 5

**Cleanliness** - עמודה קטגוריאלית מספרית, דירוג מ0 עד 5 של הניקיון בטיסה

**Departure delay in minutes** -  עמודה נומרית, עיכוב בזמן ההמראה בדקות

**Arrival delay in minutes** -  עמודה נומרית, עיכוב בזמן הנחיתה בדקות

**Satisfaction** - עמודה קטגוריאלית, האם הנוסע ״מרוצה״ או ״ניטרלי/לא מרוצה״ מהטיסה (Satisfied/ neutral or dissatisfied) </div>

###<div dir=rtl> Column Transformer ו - Pipeline</div>

```python
imputation_numerical_cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
imputation_numerical_action = SimpleImputer(strategy= 'mean')


imputation_categorial_cols= ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Inflight wifi service', 'Departure/Arrival time convenient',
                             'Ease of Online booking', 'Gate location','Food and drink','Online boarding', 'Seat comfort','Inflight entertainment',
                             'On-board service','Leg room service','Baggage handling','Checkin service','Inflight service','Cleanliness']

imputation_categorial_action = SimpleImputer(strategy= 'most_frequent')
```

<div dir=rtl>

**imputation_numerical_cols** - יצירת רשימה של העמודות הנומריות

**imputation_numerical_action** - פעולת ההשלמה שאעשה לעמודות הנומריות. כרגע השתמשתי בממוצע. בהמשך אבחן אפשרויות נוספות.

**imputation_categorial_cols** - יצירת רשימה של העמודות הקטגוריאליות

**imputation_categorial_action** - פעולת ההשלמה שאעשה לעמודות הקטגוריאליות. את ההשלמה עשיתי לפי הערך הנפוץ ביותר כיוון שלא כל העמודות הקטגוריאליות הן מספריות ואם אשתמש בממוצע למשל, לעמודות שמיות זו תהיה שגיאה.

אף על פי שהצגתי קודם כי יש ערכים חסרים רק בעמודה נומרית אחת, אני בכל זאת עושה השלמת חסרים על כל העמודות כיוון שיכולים להיות ערכים חסרים בעמודות שונות בטבלת הבדיקה. שלב זה חשוב כדי להכין את המודל ולמנוע שגיאות.

</div>

```python
imputation_parts = [('numerical', imputation_numerical_action, imputation_numerical_cols),
                    ('categorical', imputation_categorial_action, imputation_categorial_cols)]
#שלב השלמת החסרים שיעבור לטרנספורמר
```

```python
cat_to_num_binary_cols = ['Gender', 'Customer Type' ,'Type of Travel']
cat_to_num_binary_action = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=2)
```

<div dir=rtl>

**cat_to_num_binary_cols** - יצירת רשימה של עמודות בינאריות (2 ערכים) שאותם אהפוך מקטגוריאליות לנומריות.

**cat_to_num_binary_action**- הפעולה שאפעיל על העמודות הבינאריות היא OrdinalEncoder שהופך את הערכים בהן ל0 ו1.

**handle_unknown='use_encoded_value', unknown_value=2** - כאשר המודל יתקל בקטגוריה בטבלת הבדיקה שאינה קיימת בטבלת האימון, הוא יהפוך אותה למספר 2 כדי שהחיזוי לא יתרסק.
</div>

```python
cat_to_num_ordinal_cols = ['Class']
cat_to_num_ordinal_action = OrdinalEncoder(categories= [['Eco', 'Eco Plus', 'Business']], handle_unknown='use_encoded_value', unknown_value=3)
```

<div dir=rtl>

**cat_to_num_ordinal_cols** - יצירת רשימה של העמודה הקטגוריאלית Class בעלת 3 ערכים עם סדר טבעי.

**cat_to_num_ordinal_action** - הפעולה שהפעלתי על עמודת זו היא OrdinalEncoder שהופך את הערכים למספריים. על מנת לוודא שהערכים אכן יהיו ממוספרים לפי הסדר הטבעי שלהם העברתי את הפרמטר categories ולו העברתי את הערכים בעמודה לפי סדרם.

**handle_unknown='use_encoded_value', unknown_value=2** - כאשר המודל יתקל בקטגוריה בטבלת הבדיקה שאינה קיימת בטבלת האימון, הוא יהפוך אותה למספר 3 כדי שהחיזוי לא יתרסק.
</div>

```python
cat_to_num_parts = [('binary', cat_to_num_binary_action, cat_to_num_binary_cols),
                    ('ordinal', cat_to_num_ordinal_action, cat_to_num_ordinal_cols)]
#שלב הפיכת משתנים קטגוריאליים לנומרים שיעבור לטרנספורמר
```

```python
outliers_numerical_cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
outliers_numerical_action = Winsorizer()
```

<div dir=rtl>

**outliers_numerical_cols** - יצירת רשימה של העמודות הנומריות שבהם עלולים להיות ערכים חריגים

**outliers_numerical_action** - בשלב זה בחרתי להשתמש בwinsorizer לטפל בחריגים. למעשה winsorizer הוא קובע סף עליון ותחתון לערכים בהתבסס על אחוזונים של התפלגות הנתונים, כל ערך שנופל מתחת לסף התחתון או מעל הסף העליון נחשב לערך קיצון, והוא מוחלף בערך הסף המתאים. למשל, ערך שנמצא גדול מהערך המקסימלי מוחלף בערך המקסימלי.


</div>

```python
outliers_parts = [('numerical', outliers_numerical_action, outliers_numerical_cols)]
#שלב הטיפול בחריגים שיעבור לטרנספורמר
```

```python
standardization_numerical_cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Class', 'Inflight wifi service',
                                  'Departure/Arrival time convenient','Ease of Online booking', 'Gate location','Food and drink','Online boarding',
                                  'Seat comfort','Inflight entertainment','On-board service','Leg room service','Baggage handling',
                                  'Checkin service','Inflight service','Cleanliness']
standardization_numerical_action = StandardScaler()
```

<div dir=rtl>

**standardization_numerical_cols** - יצירת רשימה של העמודות המספריות אותן צריך לתקנן

**standardization_numerical_action** - תקנון (התחלתי) בעזרת standardscaler שמתאים את טווח הערכים כך שהממוצע יהיה 0 והסטיית התקן תהיה 1.
</div>

```python
standardization_parts = [('numerical', standardization_numerical_action, standardization_numerical_cols )]
#שלב התקנון שיעבור לטרנספורמר
```

```python
imputation_action = ColumnTransformer(transformers= imputation_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs= -1)
cat_to_num_action = ColumnTransformer(transformers= cat_to_num_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs=-1)
outliers_action = ColumnTransformer(transformers= outliers_parts, remainder= 'passthrough', verbose_feature_names_out=False, n_jobs=-1)
standardization_action = ColumnTransformer(transformers= standardization_parts, remainder= 'passthrough',verbose_feature_names_out=False, n_jobs=-1)
```

<div dir=rtl>

יצירת הcolumntransformer לפי סדר הפעולות שכתבתי מעלה, והכנת שלבי העיבוד המקדים שיעבור לפייפליין.

**verbose_feature_names_out=False** - הפרמטר verbose_feature_names_out=False קובע שהשמות של המשתנים שיצאו מתהליך עיבוד הנתונים יהיו קצרים ופשוטים, ללא תוספות או מידע נוסף על המקור או התהליכים שעברו עליהם. למשל כאשר השלמנו חסרים על העמודה flight distance  אז שם המשתנה המושלם ישאר ולא יהפוך ל - flight_distance__imputed. כלומר זה ישאר כך: imputation_numerical_cols.זה נחוץ כיוון שאחרת אלו לא יהיו אותם שמות כמו טבלת הקלט ואז תהיה שגיאה כי העמודה אינה קיימת בטבלת הבדיקה.
</div>

```python
preprocessing_steps = [('impute', imputation_action),
                       ('cat_to_num', cat_to_num_action),
                       ('outliers', outliers_action),
                       ('standardize', standardization_action)]
 #הגדרת השלבים הכנה לפייפליין. ראשית, השלמת ערכים חסרים, לאחר מכן הפיכת עמודות קטגוריאליות לנומריות, לאחר מכן טיפול בחריגים ולבסוף תקנון
```

```python
preprocessor = Pipeline(steps= preprocessing_steps)
#הגדרת שלבי העיבוד המקדים בתוך פייפליין
```
