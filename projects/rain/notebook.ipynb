# עיבוד נתונים

```python
#החזרת טבלה כל פעם
lrn.set_config(transform_output = "pandas")
```

```python
#העמודות הקטגוריאליות
cat_hc_cols = ["Location"] #הרבה קטגוריות
cat_lc_cyc_cols = ['WindGustDir', 'WindDir9am', 'WindDir3pm'] #קטגוריות מעגליות
cat_lc_bin_cols = ["RainToday"] #קטגוריות בינאריות
cat_lc_cols = cat_lc_cyc_cols + cat_lc_bin_cols #מעט קטגוריות


#העמודות הנומריות
num_cols = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
            'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
            'Pressure9am', 'Pressure3pm', 'Cloud9am','Cloud3pm', 'Temp9am', 'Temp3pm']
```

```python
#פעולות מילוי חסרים
impute_num_action = SimpleImputer() #עמודות נומריות - ממולאות בממוצע (ברירת המחדל)
impute_cat_lc_action = SimpleImputer(strategy  = 'most_frequent') #עמודות קטגוריאליות, מעט קטגוריות - ממולאות בערך הכי נפוץ
impute_cat_hc_action = SimpleImputer(strategy = "constant", fill_value = "missing") #'missing' עמודות קטגוריאליות, הרבה קטגוריות - מילוי הערך

#פעולות המרת עמודות קטגוריאליות למספריות
cat_to_num_hc_action = TargetEncoder(random_state = 42) #הרבה קטגוריות - החלפה של ערך מסוים בעמודה לפי אחוז השורות בעמודה המוחלפת שמתויגות בעמודת המטרה עם ערך מסוים
cat_to_num_lc_bin_action = OrdinalEncoder(handle_unknown = "use_encoded_value", unknown_value = 2) #קטגוריות בינאריות - שימוש בפונקציה שמבצעת החלפה לפי סדר טבעי למספרים, 0 ו-1 במקרה הזה
```

```python
#הערכים של העמודות הקטגוריאליות עם קטגוריות מעגליות
windrose = ['N', 'NNE', "NE", 'ENE',
            'E', 'ESE', 'ES', 'SSE',
            'S', 'SSW', 'SW', 'WSW',
            'W', 'WNW', 'NW', 'NNW']

#המרה לערכים מספריים
cat_to_num_cyc_to_num_action = OrdinalEncoder(categories = [windrose] * len(cat_lc_cyc_cols),
                                              handle_unknown = "use_encoded_value",
                                              unknown_value = -1)

#סידור קשר מעגלי
cat_to_num_cyc_cycle_action = helpers.make_cyclic_transformer(len(windrose))
```

```python
#פעולת ההמרה הכוללת של העמודות הקטגוריאליות המעגליות לנומריות
cat_to_num_lc_cyc_action = Pipeline(steps = [('to_num', cat_to_num_cyc_to_num_action),
                                             ("cycle", cat_to_num_cyc_cycle_action)])
```

```python
#הפעולה הכוללת של מילוי החסרים, מורכבת מפעולות שהוצגו בתאים קודמים
impute_action = ColumnTransformer(transformers = [("num", impute_num_action, num_cols),
                                                  ('cat_lc', impute_cat_lc_action, cat_lc_cols),
                                                  ("cat_hc", impute_cat_hc_action, cat_hc_cols)],
                                  remainder = "passthrough",
                                  n_jobs = -1,
                                  verbose_feature_names_out = False)

#הפעולה הכוללת של העברת העמודות הקטגוריאליות לנומריות, מורכבת מפעולות שהוצגו בתאים קודמים
cat_to_num_action = ColumnTransformer(transformers = [("hc", cat_to_num_hc_action, cat_hc_cols),
                                                      ('lc_bin', cat_to_num_lc_bin_action, cat_lc_bin_cols),
                                                      ("lc_cyc", cat_to_num_lc_cyc_action, cat_lc_cyc_cols)],
                                      remainder = "passthrough",
                                      n_jobs = -1,
                                      verbose_feature_names_out = False)

#הכי קרוב whisker-פעולת הטיפול בחריגים - החלפה של ערכים חריגים ב
outliers_action = Winsorizer()

#פעולת הנרמול - נרמול הנתונים (שעכשיו כולם מספריים) בין 0 ו-1
scale_action = MinMaxScaler()
```

```python
#התהליך הכולל שהטבלה תעבור
pre_processing = Pipeline([
    ('impute', impute_action),
    ('cat_to_num',cat_to_num_action),
    ('outliers', outliers_action),
    ('scale', scale_action),
])
```

```python
#עיבוד טבלת האימון והצגתה
preprocessed_data = pre_processing.fit_transform(X_train, y_train)
preprocessed_data
```

```python
preprocessed_data.corr() #יצירת טבלת קורלציה
fig, ax = plt.subplots(figsize=(20, 15)) #שינוי גודל
sns.heatmap(preprocessed_data.corr(), annot = True, ax = ax); #יצירת מפת חום לטבלת הקורלציה
```

<div dir=rtl>
רוב הטבלה צבועה בצבעי סגול ואדום, בערך בטווח הערכים 0.2- עד 0.3, ויש קצת משבצות יותר כהות וקצת משבצות יותר בהירות. באלכסון ה-1 שנוצר כי כל עמודה היא בקורלציה מושלמת עם עצמה יש סביב חלק מהמשבצות כמה משבצות יותר כתומות, שמצביעות על קורלציה יותר גבוהה. הן נובעות מכך שנבדקות עמודות עם קטגוריות דומות, למשל הרוח בבוקר ובצהריים. יש כמה מבצות מאוד כהות נוצרות מכך תנאים מסוימים פחות קורים ביחד בחיים האמיתיים, לדוגמה שמיים מכוסים עננים לרוב לא באים עם הרבה שמש. נראה שבין רוב הזוגות של ערכים אין קורלציה גבוהה ולכן כנראה בחיזוי יהיה צורך להסתמך על הערכים בתור מה שהם מייצגים כאשר הם ביחד יותר ממשמעותם בפני עצמם.
</div>

# למידת מכונה

```python
#הגדרת פייפליין
clf = Pipeline([
    ('pre_processor', pre_processing),
    ('classifier', KNeighborsClassifier())
])

#הגדרת פרמטרים
params = [
    {
        'pre_processor__scale': [MinMaxScaler(), StandardScaler(), RobustScaler(), MaxAbsScaler()],
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [10, 20, 30]
    },
    {
        'pre_processor__scale': [MinMaxScaler(), StandardScaler(), RobustScaler(), MaxAbsScaler()],
        'classifier': [SVC(random_state = 42)],
        'classifier__C': [1, 10, 15],
        'classifier__kernel': ['linear', 'rbf', 'poly', 'sigmoid']
    }
]

scorer = make_scorer(recall_score)

#GridSearch-הגדרת ה
gs = GridSearchCV(estimator = clf,
                  param_grid = params,
                  scoring = scorer,
                  n_jobs = -1,
                  verbose= 10,
                  error_score= 'raise')
```
