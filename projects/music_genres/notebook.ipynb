## נרמול נתונים

```python
lrn.set_config(transform_output= 'pandas')
```

```python
str_to_float_cols = ['tempo']                                  # נחלק את העמודות לקבוצות לפי התוכן שלהם
num_cols = ['popularity', 'acousticness', 'danceability',
            'duration_ms', 'energy', 'instrumentalness',
            'liveness', 'loudness', 'speechiness',
            'tempo', 'valence']
ord_cat_cols = ['mode','obtained_date']
hc_cols = ['artist_name', 'track_name', 'key']
```

```python
conversion_str_to_float_action = FunctionTransformer(
                                    lambda df: df.apply(
                                        func= lambda s: pd.to_numeric(arg= s,    # ניצור את תהליכי הנרמול שלנו לפי סוג העמודות
                                                            errors= 'coerce'),
                                        axis= 1))
conversion_str_to_date_action = FunctionTransformer(
                                  lambda df: pd.to_datetime(df))
imputation_num_action = SimpleImputer()
imputation_cat_action = SimpleImputer(strategy= 'most_frequent')
imputation_hc_action = SimpleImputer(strategy= 'constant',
                                     missing_values= 'missing')
cat_to_num_ord_action = OrdinalEncoder(handle_unknown= 'use_encoded_value',
                                       unknown_value= -1)
cat_to_num_hc_action = TargetEncoder(random_state= 42)
```

```python
conversion_parts = [('str_to_float',
                     conversion_str_to_float_action,
                     str_to_float_cols)]
imputation_parts = [('num', imputation_num_action, num_cols), # ניצור את תהליכי האימפיוטציה
                    ('cat', imputation_cat_action, ord_cat_cols),
                    ('hc', imputation_hc_action, hc_cols)]
cat_to_num_parts = [('ord', cat_to_num_ord_action, ord_cat_cols),
                    ('hc', cat_to_num_hc_action, hc_cols)]
```

```python
conversion_action = ColumnTransformer(transformers= conversion_parts,      #הכנסת העמודות אל תוך הקולום טרנספורמר
                                      remainder= 'passthrough',
                                      n_jobs = -1,
                                      verbose_feature_names_out= False)
imputation_action = ColumnTransformer(transformers= imputation_parts,
                                      remainder= 'passthrough',
                                      n_jobs = -1,
                                      verbose_feature_names_out= False)
cat_to_num_action = ColumnTransformer(transformers= cat_to_num_parts,
                                      remainder= 'passthrough',
                                      n_jobs = -1,
                                      verbose_feature_names_out= False)
outliers_action = Winsorizer()
standardization_action = MinMaxScaler()
```

```python
preprocessing_steps = [('convert', conversion_action),         # נחבר הכל ביחד
                       ('impute', imputation_action),
                       ('cat_to_num', cat_to_num_action),
                       ('outliers', outliers_action),
                       ('standardize', standardization_action)]
```

```python
preprocessor = Pipeline(steps= preprocessing_steps)
```

```python
preprocessor
```

```python
preprocessor.fit_transform(X_train, y_train) # יישום הקוד בעזרת הטבלה שלנו
```

# למידת מכונה

<div dir="rtl">
לשם אימון המודל בחרתי לשמור את כל העמודות משום שכל אחת מן העמודות חיונית לחיזוי ותעזור לנו לקבוע את סגנון המוזיקה של כל שיר.
<div/>

```python
# הגדרת הפייפליין
clf = Pipeline([
    ('pre_processor', preprocessor),
    ('classifier', KNeighborsClassifier())
])

# הגדרת הפרמטרים
params = [
    {
        'pre_processor__standardize': [MinMaxScaler(), StandardScaler(), RobustScaler(), MaxAbsScaler()],
        'classifier': [KNeighborsClassifier()],
        'classifier__n_neighbors': [1,2,3,4]
    },
    {
        'pre_processor__standardize': [MinMaxScaler(), StandardScaler(), RobustScaler(), MaxAbsScaler()],
        'classifier': [SVC(random_state= 42)],
        'classifier__C': [1, 2, 3],
        'classifier__kernel': ['linear', 'rbf','poly']
    }
]

scorer = make_scorer(accuracy_score)

gs = GridSearchCV(estimator = clf,
                  param_grid = params,
                  scoring = scorer,
                  n_jobs = -1,
                  verbose= 10,
                  error_score='raise')
```

```python
gs = helpers.fit(gs, X_train, y_train)
```

The fitting ran for almost 3 hours.

```python
gs.best_params_ # מציאת ההיפר פרמטרים הכי טובים
```

```python
y_predict = helpers.predict(gs, X_test)
```

# חקר ובחינת ביצועים
